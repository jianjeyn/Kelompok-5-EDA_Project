Pariwisata merupakan salah satu sektor yang memiliki peran penting dalam perekonomian daerah, termasuk di Provinsi Jawa Barat. Dengan keindahan alam, warisan budaya, dan atraksi wisata lainnya, Jawa Barat menarik banyak wisatawan, baik domestik maupun internasional. Pengembangan sektor pariwisata dapat memberikan dampak positif terhadap perekonomian masyarakat, seperti peningkatan pendapatan daerah, penciptaan lapangan kerja, dan pertumbuhan usaha mikro, kecil, dan menengah (UMKM).

Namun, untuk memahami secara lebih mendalam bagaimana pengembangan pariwisata benar-benar memengaruhi ekonomi masyarakat Jawa Barat, diperlukan analisis lebih lanjut terhadap berbagai faktor yang terlibat. Salah satu metode yang dapat digunakan adalah Exploratory Data Analysis (EDA), yang memungkinkan kita untuk mengeksplorasi data dari berbagai dimensi.

Dengan memanfaatkan data dari berbagai sumber, seperti Badan Pusat Statistik (BPS) dan Open Data Jabar, analisis ini berusaha menjawab beberapa pertanyaan penting:

1. Bagaimana pengaruh jumlah pengunjung objek wisata terhadap Produk Domestik Regional Bruto (PDRB) di Jawa Barat?
2. Apakah pengembangan sektor pariwisata berhubungan dengan peningkatan jumlah UMKM di berbagai kabupaten/kota?
3. Bagaimana pengaruh perkembangan jumlah tenaga kerja dibidang pariwisata terhadap tingkat pengangguran terbuka?
4. Bagaimana dampak keseluruhan dari pengembangan pariwisata terhadap ekonomi masyarakat di Jawa Barat?

## **Filtering data**

Melakukan proses penyaringan data.

###Filtering Dataset Obyek Wisata

Proses penyaringan data dimulai dengan mengambil data yang hanya mencakup tahun 2018 hingga 2022 untuk fokus pada periode yang relevan. Kemudian, kolom-kolom seperti id, kode provinsi, nama provinsi, dan satuan dihapus karena dianggap kurang relevan untuk analisis yang hanya berfokus pada wilayah Jawa Barat. Selain itu, nama kolom yang memuat jumlah Obyek Daya Tarik Wisata diubah untuk memperjelas bahwa kolom tersebut berisi informasi tentang jumlah lokasi wisata, sehingga lebih mudah dipahami dalam konteks analisis.

```python
# Menampilkan data hanya dari tahun 2018 sampai tahun 2022
obyekwisata = obyekwisata[obyekwisata['tahun'].between(2018, 2022)]

# Menghapus kolom id, kode_provinsi, nama_provinsi, dan satuan karena tingkat korelasi yang tinggi dengan dataset lainnya
obyekwisata = obyekwisata.drop(['nama_provinsi', 'kode_provinsi', 'id','satuan'], axis=1)

# Untuk memperjelas satuan, maka kolom jumlah_lokasi direplace menjadi jumlah_lokasi_odtw
obyekwisata = obyekwisata.rename(index=str, columns={"jumlah_odtw": "jumlah_lokasi_odtw"})

# Menampilkan DataFrame hasil
obyekwisata.head(5)
```

Selain itu, nama kolom yang memuat jumlah Tenaga Kerja diubah untuk memperjelas satuan yang digunakan.

###Filtering Dataset PDRB ADHK (Produk Domestik Regional Bruto Atas Dasar Harga Konstan)

Nilai pada tahun 2018 dianggap anomali sehingga diubah menjadi nilai kosong (NaN).

Setelah itu, dilakukan interpolasi linier berdasarkan data dari tahun-tahun sebelumnya (2013-2017) untuk memperkirakan nilai yang hilang. Hasil interpolasi kemudian digabungkan kembali ke dalam dataset asli untuk menggantikan nilai pada tahun 2018. Setelah semua nilai diperbarui, data difilter hanya untuk tahun 2018 hingga 2022, dan tipe data pada kolom pdrb_adhk_miliar diubah dari float menjadi integer untuk memastikan format data sesuai dengan analisis lebih lanjut.

```python
# Menghapus kolom id, kode_provinsi, nama_provinsi, dan satuan karena kurang relevan
pdrb_adhk = pdrb_adhk.drop(['nama_provinsi', 'kode_provinsi', 'id','satuan'], axis = 1)
pdrb_adhk = pdrb_adhk.rename(index=str, columns={"pdrb_adhk": "pdrb_adhk_miliar"})


# Ganti semua nilai di tahun 2018 dengan NaN
pdrb_adhk.loc[pdrb_adhk['tahun'] == 2018, 'pdrb_adhk_miliar'] = pd.NA

# Lakukan interpolasi linier berdasarkan tahun-tahun sebelumnya (2013-2017)
interpolated_values = (
    pdrb_adhk.groupby('nama_kabupaten_kota')['pdrb_adhk_miliar']
    .apply(lambda group: group.interpolate(method='linear'))
)

# Reset indeks dari interpolated_values
interpolated_values = interpolated_values.reset_index(level=0, drop=True)

# Gabungkan kembali hasil interpolasi ke dalam DataFrame asli
pdrb_adhk['pdrb_adhk_miliar'] = interpolated_values
```

###Filtering Data Jumlah UMKM

Proses modifikasi data jumlah_umkm dimulai dengan mendefinisikan fungsi modify_nama_kabupaten_kota untuk mengubah setiap nama kabupaten/kota menjadi huruf kapital. Jika nama tersebut tidak mengandung kata "KOTA", maka fungsi ini menambahkan prefiks "KABUPATEN" di depan nama. Setelah itu, fungsi ini diterapkan pada kolom nama_kabupaten_kota dalam dataset jumlah_umkm untuk memastikan konsistensi penamaan.

```python
# Fungsi untuk mengubah nama_kabupaten_kota
def modify_nama_kabupaten_kota(nama):
    # Ubah menjadi huruf besar
    nama = nama.upper()

    # Cek apakah ada kata "KOTA" di dalam nama
    if "KOTA" not in nama:
        # Tambahkan "KABUPATEN" di depan
        nama = "KABUPATEN " + nama

    return nama

# Terapkan fungsi ke kolom 'nama_kabupaten_kota'
jumlah_umkm['nama_kabupaten_kota'] = jumlah_umkm['nama_kabupaten_kota'].apply(modify_nama_kabupaten_kota)

# Mengubah tipe data kolom 'tahun' di jumlah_umkm menjadi integer
jumlah_umkm['tahun'] = jumlah_umkm['tahun'].astype(int)

# Menampilkan DataFrame hasil
jumlah_umkm.head(5)
```

###Data Distribusi Jumlah Obyekwisata Tiap Kabupaten/Kota di Jawa Barat

```python
# Mengelompokkan dan menjumlahkan lokasi obyek wisata per kabupaten/kota dan tahun
data_aggregated = obyekwisata.groupby(['nama_kabupaten_kota', 'tahun'])['jumlah_lokasi_odtw'].sum().reset_index()

# Mengurutkan kabupaten/kota berdasarkan rata-rata jumlah lokasi obyek wisata
average_locations = obyekwisata.groupby('nama_kabupaten_kota')['jumlah_lokasi_odtw'].mean().reset_index()
sorted_kabupaten = average_locations.sort_values(by='jumlah_lokasi_odtw', ascending=False)['nama_kabupaten_kota']

# Mengurutkan sumbu x berdasarkan jumlah rata-rata lokasi obyek wisata
data_aggregated['nama_kabupaten_kota'] = pd.Categorical(data_aggregated['nama_kabupaten_kota'],
    categories=sorted_kabupaten, ordered=True)

# Membuat bar chart
plt.figure(figsize=(14, 8))  # Ukuran grafik yang lebih besar untuk memuat semua data
sns.barplot(data=data_aggregated, x='nama_kabupaten_kota', y='jumlah_lokasi_odtw', hue='tahun')

# Menambahkan label dan judul
plt.xlabel('Nama Kabupaten/Kota')
plt.ylabel('Jumlah Lokasi Obyek Wisata')
plt.title('Perbandingan Jumlah Lokasi Obyek Wisata per Kabupaten/Kota Setiap Tahun')
plt.xticks(rotation=90)  # Memutar label x untuk keterbacaan
plt.legend(title='Tahun')
plt.tight_layout()  # Menyesuaikan layout agar tidak terpotong

# Menampilkan grafik
plt.show()
```

### Korelasi Antara PDRB dengan Jumlah Pengunjung

```python
# # Step 1: Average `pdrb_adhk` over the years for each `kode_kabupaten_kota` and `tahun`
avg_pdrb = pdrb_adhk.groupby(['nama_kabupaten_kota', 'tahun']).agg(
    avg_pdrb=('pdrb_adhk_miliar', 'sum')
).reset_index()

# Step 2: Average `jumlah_orang_pengunjung` over the years for each `kode_kabupaten_kota` and `tahun`
avg_jumlah_pengunjung = jumlah_pengunjung.groupby(['nama_kabupaten_kota','tahun']).agg(
    avg_jumlah_orang_pengunjung=('jumlah_orang_pengunjung', 'sum')
).reset_index()

avg_jumlah_pengunjung['avg_jumlah_orang_pengunjung'] = avg_jumlah_pengunjung['avg_jumlah_orang_pengunjung'] / 1e6


# Step 3: Merge the averaged data for PDRB
merged_avg_df = pd.merge(avg_pdrb, avg_jumlah_pengunjung, on=['nama_kabupaten_kota', 'tahun'])

# Step 5: Create scatter plots for each year with 2 charts per row
unique_years = merged_avg_df['tahun'].unique()
n_years = len(unique_years)

# Calculate number of rows needed for 2 charts per row
n_rows = (n_years + 1) // 2  # Ceiling division

# Create subplots
fig, axes = plt.subplots(n_rows, 2, figsize=(15, 5 * n_rows))
axes = axes.flatten()  # Flatten the 2D array of axes for easy indexing

for i, year in enumerate(unique_years):
    sns.regplot(
        data=merged_avg_df[merged_avg_df['tahun'] == year],
        x='avg_jumlah_orang_pengunjung',
        y='avg_pdrb',
        ax=axes[i],
        scatter_kws={"s": 50, "alpha": 0.7},
        line_kws={"color": "red"}
    )
    axes[i].set_title(f"Korelasi PDRB ADHK dengan Jumlah Pengunjung Wisata (Tahun {year})")
    axes[i].set_xlabel("Jumlah Pengunjung Wisata (juta)")
    axes[i].set_ylabel("PDRB ADHK")

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

# Adjust layout
plt.tight_layout()
plt.show()
```

###Korelasi Antara Jumlah Obyek Wisata Dengan Jumlah Pengunjung

```python

# Menghitung jumlah setiap jenis ODTW per kabupaten/kota dan per tahun
total_odtw_per_kabupaten = obyekwisata.groupby(
    ['kode_kabupaten_kota', 'nama_kabupaten_kota', 'tahun']
)['jumlah_lokasi_odtw'].sum().reset_index()

# Mengatur tahun yang akan diproses
years = [2018, 2019, 2020, 2021, 2022]

# Membuat subplots
fig, axes = plt.subplots(3, 2, figsize=(14, 12))  # 3 rows, 2 columns
axes = axes.flatten()  # Flatten untuk memudahkan indexing

# Loop untuk setiap tahun
for i, year in enumerate(years):
    # Filter data untuk tahun tertentu
    obyekwisata_year = total_odtw_per_kabupaten[total_odtw_per_kabupaten['tahun'] == year]
    jumlah_pengunjung_year = jumlah_pengunjung[jumlah_pengunjung['tahun'] == year]

    # Menghitung total jumlah pengunjung per kabupaten/kota
    total_pengunjung = jumlah_pengunjung_year.groupby('kode_kabupaten_kota')['jumlah_orang_pengunjung'].sum().reset_index()

    # Menggabungkan kedua DataFrame
    merged_data = pd.merge(obyekwisata_year, total_pengunjung, on='kode_kabupaten_kota', how='inner')
    merged_data.rename(columns={
        'jumlah_lokasi_odtw': 'Total Obyek Wisata',
        'jumlah_orang_pengunjung': 'Total Pengunjung'
    }, inplace=True)

    # Membuat scatter plot di subplot
    sns.scatterplot(data=merged_data, x='Total Obyek Wisata', y='Total Pengunjung', ax=axes[i])
    sns.regplot(data=merged_data, x='Total Obyek Wisata', y='Total Pengunjung', scatter=False, color='red', ax=axes[i])

    # Menambahkan label dan judul
    axes[i].set_title(f'Korelasi antara Jumlah Obyek Wisata dan Jumlah Pengunjung ({year})')
    axes[i].set_xlabel('Total Obyek Wisata')
    axes[i].set_ylabel('Total Pengunjung (juta)')
    axes[i].grid()

# Menghapus subplot kosong jika ada (misal di tahun genap)
if len(years) < len(axes):
    for j in range(len(years), len(axes)):
        fig.delaxes(axes[j])

# Menyesuaikan layout
plt.tight_layout()
plt.show()
```

Kelima chart diatas menggambarkan bagaimana korelasi antara jumlah pengunjung wisata dengan total obyek wisata dari tahun 2018-2022. Setiap titik pada chart menunjukan data dari tiap kabupaten atau kota di Jawa Barat.

Berdasarkan chart diatas, bisa dilihat bahwa semakin banyak obyek wisata maka, total pengunjungnya pun ikut semakin banyak. Hal ini menunjukkan bahwa kedua data tersebut memiliki korelasi yang tinggi, korelasi ini sangat relevan terhadap topik utama analisis, yaitu pengaruh pengembangan pariwisata terhadap ekonomi masyarakat Jawa Barat. Peningkatan jumlah pengunjung sebagai hasil dari bertambahnya objek wisata menunjukkan potensi pertumbuhan ekonomi daerah melalui sektor pariwisata. Dengan bertambahnya wisatawan, peluang peningkatan pendapatan masyarakat lokal juga ikut naik, terutama bagi pelaku usaha mikro, kecil, dan menengah (UMKM) yang bisa memanfaatkan arus wisatawan untuk menjual produk atau layanan mereka. ### Korelasi Jumlah Pengunjung dan Jumlah Pengunjung Akomodasi Wisata

```python
# 1. Menghitung total jumlah pengunjung per kabupaten/kota dan tahun untuk dataset jumlah_pengunjung
total_pengunjung = jumlah_pengunjung.groupby(['nama_kabupaten_kota', 'tahun'])['jumlah_orang_pengunjung'].sum().reset_index()

# 2. Menghitung rata-rata jumlah pengunjung per kabupaten/kota dari 5 tahun (dataset jumlah_pengunjung)
rata_pengunjung = total_pengunjung.groupby('nama_kabupaten_kota')['jumlah_orang_pengunjung'].mean().reset_index()

# 3. Menghitung total jumlah pengunjung per kabupaten/kota dan tahun untuk dataset jumlah_pengunjung_akomodasi_wisata
total_pengunjung_akomodasi = jumlah_pengunjung_akomodasi_wisata.groupby(['nama_kabupaten_kota', 'tahun'])['jumlah_orang_pengunjung'].sum().reset_index()

# 4. Menghitung rata-rata jumlah pengunjung per kabupaten/kota dari 5 tahun (dataset jumlah_pengunjung_akomodasi_wisata)
rata_pengunjung_akomodasi = total_pengunjung_akomodasi.groupby('nama_kabupaten_kota')['jumlah_orang_pengunjung'].mean().reset_index()

# 5. Menggabungkan rata-rata dataset berdasarkan nama kabupaten/kota
merged_data = pd.merge(rata_pengunjung, rata_pengunjung_akomodasi, on='nama_kabupaten_kota', suffixes=('_wisata', '_akomodasi'))

# merged_data

# 6. Membuat scatter plot dengan garis tren
plt.figure(figsize=(12, 6))

# Scatter plot
sns.scatterplot(data=merged_data, x='jumlah_orang_pengunjung_wisata', y='jumlah_orang_pengunjung_akomodasi', s=100, color='blue')

# Garis tren (regresi linear)
sns.regplot(data=merged_data, x='jumlah_orang_pengunjung_wisata', y='jumlah_orang_pengunjung_akomodasi', scatter=False, color='red')

# Mengatur judul dan label
plt.title('Korelasi antara Jumlah Pengunjung Wisata dan Jumlah Pengunjung Akomodasi Wisata per Kabupaten/Kota')
plt.xlabel('Rata-rata Jumlah Pengunjung Wisata (juta)')
plt.ylabel('Rata-rata Jumlah Pengunjung Akomodasi Wisata (juta)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Berdasarkan grafik di atas, hubungan antara rata-rata jumlah pengunjung wisata dengan rata-rata jumlah pengunjung akomodasi wisata per kabupaten/kota menunjukkan korelasi yang sangat lemah, seperti yang diindikasikan oleh garis tren yang hampir datar. Meskipun ada beberapa kabupaten/kota dengan jumlah pengunjung wisata yang lebih tinggi, peningkatan ini tidak secara langsung berhubungan kuat dengan peningkatan jumlah pengunjung akomodasi wisata. Secara keseluruhan, grafik ini mengindikasikan bahwa tingginya jumlah pengunjung ke objek wisata tidak selalu diiringi dengan peningkatan yang sebanding pada jumlah pengunjung di akomodasi wisata. ```python

```python
# 1. Menghitung total jumlah pengunjung per kabupaten/kota dan tahun untuk dataset jumlah_pengunjung_akomodasi_wisata
total_pengunjung_akomodasi = jumlah_pengunjung_akomodasi_wisata.groupby(['nama_kabupaten_kota', 'tahun'])['jumlah_orang_pengunjung'].sum().reset_index()

# 2. Menghitung rata-rata jumlah pengunjung per kabupaten/kota dari 5 tahun (dataset jumlah_pengunjung_akomodasi_wisata)
rata_pengunjung_akomodasi = total_pengunjung_akomodasi.groupby('nama_kabupaten_kota')['jumlah_orang_pengunjung'].mean().reset_index()

# 3. Menghitung total jumlah unit hotel per kabupaten/kota dan tahun untuk dataset hotel
total_hotel = hotel.groupby(['nama_kabupaten_kota', 'tahun'])['jumlah_unit_hotel'].sum().reset_index()

# 4. Menghitung rata-rata jumlah unit hotel per kabupaten/kota dari 5 tahun (dataset hotel)
rata_hotel = total_hotel.groupby('nama_kabupaten_kota')['jumlah_unit_hotel'].mean().reset_index()

# 5. Menggabungkan rata-rata dataset berdasarkan nama kabupaten/kota
merged_data = pd.merge(rata_pengunjung_akomodasi, rata_hotel, on='nama_kabupaten_kota')

# 6. Memisahkan data: satu tanpa 5 data entry tertinggi dan satu dengan semua data
outlier = merged_data.sort_values(by='jumlah_orang_pengunjung', ascending=False).iloc[1:]
all_data = merged_data  # Menggunakan semua data tanpa pengecualian

# 7. Membuat scatter plot bersebelahan
fig, axes = plt.subplots(1, 2, figsize=(18, 6), sharey=True)

# Scatter plot untuk data tanpa 5 terbesar
sns.scatterplot(data=outlier, x='jumlah_unit_hotel', y='jumlah_orang_pengunjung', s=100, color='blue', ax=axes[1])
sns.regplot(data=outlier, x='jumlah_unit_hotel', y='jumlah_orang_pengunjung', scatter=False, color='red', ax=axes[1])
axes[1].set_title('Korelasi Tanpa Outlier')
axes[1].set_xlabel('Rata-rata Jumlah Unit Hotel')
axes[1].set_ylabel('Rata-rata Jumlah Pengunjung Akomodasi Wisata')

# Scatter plot untuk semua data
sns.scatterplot(data=all_data, x='jumlah_unit_hotel', y='jumlah_orang_pengunjung', s=100, color='purple', ax=axes[0])
sns.regplot(data=all_data, x='jumlah_unit_hotel', y='jumlah_orang_pengunjung', scatter=False, color='green', ax=axes[0])
axes[0].set_title('Korelasi dengan Semua Data')
axes[0].set_xlabel('Rata-rata Jumlah Unit Hotel')

plt.suptitle('Perbandingan Jumlah Unit Hotel dan Pengunjung Akomodasi Wisata per Kabupaten/Kota')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

Namun, ada beberapa outlier yang memberikan dampak signifikan terhadap analisis ini. Setelah menghapus outlier tersebut dari grafik, hubungan antara jumlah unit hotel dan pengunjung akomodasi wisata menjadi sangat lemah. Hal ini diperkirakan disebabkan oleh berbagai faktor. Salah satunya adalah variasi ukuran hotel; beberapa hotel memiliki kapasitas besar, sehingga jumlah hotel yang ada tidak selalu mencerminkan total pengunjung akomodasi yang sebenarnya.

Faktor lain yang perlu diperhatikan adalah rentang waktu yang ditampilkan dalam grafik, yaitu hanya mencakup periode dari 2018 hingga 2022, yang terbilang singkat untuk mengamati dinamika industri perhotelan. Proses pembangunan hotel memerlukan waktu yang tidak singkat, mulai dari perencanaan hingga konstruksi, serta pengesahan izin yang kompleks. Selain itu, industri perhotelan adalah bisnis yang penuh tantangan dan risiko, di mana keberhasilan sebuah hotel tidak hanya bergantung pada jumlah unit yang tersedia, tetapi juga pada faktor-faktor lain seperti lokasi, dan strategi pemasarannya. Dengan mempertimbangkan aspek-aspek tersebut, kita dapat lebih memahami mengapa korelasi antara jumlah hotel dan jumlah pengunjung akomodasi wisata bisa sangat bervariasi.

Hal ini masuk akal karena objek wisata dapat menarik lebih banyak pengunjung, yang pada gilirannya menciptakan peluang pasar lebih besar bagi UMKM di sekitar area wisata. Korelasi positif antara jumlah objek wisata dan UMKM ini berdampak pada topik utama yaitu, dengan meningkatnya jumlah objek wisata UMKM memiliki kesempatan untuk tumbuh dan berkembang, yang berpotensi meningkatkan pendapatan dan kesejahteraan masyarakat lokal. ### Heatmap

```python
obyekwisata = obyekwisata.groupby(['nama_kabupaten_kota', 'tahun']).agg(
    jumlah_lokasi_odtw=('jumlah_lokasi_odtw', 'sum')
).groupby('nama_kabupaten_kota').agg(
    jumlah_lokasi_odtw=('jumlah_lokasi_odtw', 'mean')
).reset_index()

tenaga_kerja_pariwisata = tenaga_kerja_pariwisata.groupby(['nama_kabupaten_kota', 'tahun']).agg(
    jumlah_orang_tenaga_kerja=('jumlah_orang_tenaga_kerja', 'sum')
).groupby('nama_kabupaten_kota').agg(
    jumlah_orang_tenaga_kerja=('jumlah_orang_tenaga_kerja', 'mean')
).reset_index()

jumlah_pengunjung = jumlah_pengunjung.groupby(['nama_kabupaten_kota', 'tahun']).agg(
    jumlah_orang_pengunjung=('jumlah_orang_pengunjung', 'sum')
).groupby('nama_kabupaten_kota').agg(
    jumlah_orang_pengunjung=('jumlah_orang_pengunjung', 'mean')
).reset_index()

pdrb_adhk = pdrb_adhk.groupby(['nama_kabupaten_kota']).agg(
    pdrb_adhk_miliar=('pdrb_adhk_miliar', 'mean')
).reset_index()

tingkat_pengangguran_terbuka = tingkat_pengangguran_terbuka.groupby(['nama_kabupaten_kota']).agg(
    tingkat_persen_pengangguran_terbuka=('tingkat_persen_pengangguran_terbuka', 'mean')
).reset_index()

hotel = hotel.groupby(['nama_kabupaten_kota']).agg(
    jumlah_unit_hotel=('jumlah_unit_hotel', 'mean')
).reset_index()

jumlah_pengunjung_akomodasi_wisata = jumlah_pengunjung_akomodasi_wisata.groupby(['nama_kabupaten_kota', 'tahun']).agg(
    jumlah_orang_pengunjung=('jumlah_orang_pengunjung', 'sum')
).groupby('nama_kabupaten_kota').agg(
    jumlah_orang_pengunjung_=('jumlah_orang_pengunjung', 'mean')
).reset_index()

jumlah_umkm = jumlah_umkm.groupby(['nama_kabupaten_kota']).agg(
    jumlah_umkm=('jumlah_umkm', 'mean')
).reset_index()

pdrb_per_kapita = pdrb_per_kapita.groupby(['nama_kabupaten_kota']).agg(
    pdrb_per_capita=('pdrb_per_capita', 'mean')
).reset_index()

jumlah_pengunjung_akomodasi_wisata = jumlah_pengunjung_akomodasi_wisata.rename(
    columns={'jumlah_orang_pengunjung_': 'jumlah_pengujung_akomodasi'}
)

obyekwisata = obyekwisata.rename(
    columns={'jumlah_lokasi_odtw': 'jumlah_obyek_wisata'}
)
```

```python
merged_df = obyekwisata.merge(tenaga_kerja_pariwisata, on='nama_kabupaten_kota', how='left')
merged_df = merged_df.merge(jumlah_pengunjung, on='nama_kabupaten_kota', how='left')
merged_df = merged_df.merge(pdrb_adhk, on='nama_kabupaten_kota', how='left')
merged_df = merged_df.merge(tingkat_pengangguran_terbuka, on='nama_kabupaten_kota', how='left')
merged_df = merged_df.merge(hotel, on='nama_kabupaten_kota', how='left')
merged_df = merged_df.merge(jumlah_pengunjung_akomodasi_wisata, on='nama_kabupaten_kota', how='left')
merged_df = merged_df.merge(jumlah_umkm, on='nama_kabupaten_kota', how='left')
merged_df = merged_df.merge(pdrb_per_kapita, on='nama_kabupaten_kota', how='left')

numeric_df = merged_df.select_dtypes(include='number')

numeric_df.set_index(merged_df['nama_kabupaten_kota'], inplace=True)

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

Heatmap di atas menampilkan hubungan korelasi antara berbagai variabel terkait sektor pariwisata dan ekonomi di Jawa Barat. Variabel-variabel seperti jumlah objek wisata, jumlah tenaga kerja, jumlah pengunjung, PDRB, tingkat pengangguran, jumlah hotel, pengunjung akomodasi, jumlah UMKM, dan PDRB per kapita dianalisis untuk melihat seberapa kuat hubungan antarvariabel tersebut.

Dari heatmap ini, beberapa korelasi yang lebih signifikan terlihat antara jumlah objek wisata dengan jumlah UMKM (korelasi positif 0,72), menunjukkan bahwa daerah dengan lebih banyak objek wisata cenderung memiliki jumlah UMKM yang lebih tinggi. Korelasi positif yang cukup kuat juga terdapat antara jumlah pengunjung dan jumlah hotel (0,43), serta antara jumlah pengunjung dan PDRB ADHK (0,61), yang mengindikasikan bahwa daerah dengan lebih banyak pengunjung biasanya memiliki lebih banyak hotel dan berkontribusi lebih besar pada PDRB ADHK. Selain itu, jumlah pengunjung akomodasi menunjukkan korelasi positif yang moderat dengan PDRB per kapita (0,59), yang menunjukkan bahwa wisatawan yang menggunakan akomodasi dapat berkontribusi terhadap peningkatan PDRB per kapita.

Sebaliknya, terdapat korelasi negatif antara jumlah objek wisata dengan tingkat pengangguran (-0,36) dan antara jumlah UMKM dengan tingkat pengangguran (-0,34), yang menunjukkan bahwa daerah dengan lebih banyak objek wisata dan UMKM cenderung memiliki tingkat pengangguran yang lebih rendah. Hubungan antara PDRB ADHK dan PDRB per kapita sangat tinggi (0,82), yang dapat diharapkan mengingat keduanya merupakan indikator yang erat terkait dalam mengukur pertumbuhan ekonomi.
